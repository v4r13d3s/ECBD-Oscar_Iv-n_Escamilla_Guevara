# Explicación de cómo afecta el principio GIGO al análisis de datos y Machine Learning

El principio GIGO (Garbage In, Garbage Out) explica justamente esto: si los datos de entrada son incorrectos o están mal formateados, los resultados del análisis o de un modelo de Machine Learning también serán erróneos. No importa qué tan sofisticado sea el algoritmo, si se alimenta con datos basura, lo que se obtiene es basura. Por eso es tan importante dedicar tiempo a limpiar y transformar los datos antes de analizarlos o entrenar un modelo.

```python
# Gráfica 1: Ventas por país
ventas_pais = dfventas.copy()
ventas_pais['Monto'] = pd.to_numeric(ventas_pais['Monto'].str.replace(',', '.'), errors='coerce')
ventas_pais = ventas_pais.groupby('País')['Monto'].sum().sort_values(ascending=True)

plt.figure(figsize=(12, 7))
bars = plt.barh(ventas_pais.index, ventas_pais.values, color='steelblue')
plt.bar_label(bars, fmt='${:,.0f}', padding=5, fontsize=9)
plt.xlabel('Monto total ($)')
plt.title('Ventas totales por país')
plt.tight_layout()
plt.show()
```

```python
# Gráfica 2: Heatmap de valores nulos
plt.figure(figsize=(12, 6))
sns.heatmap(dfventas.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Heatmap de valores nulos')
plt.xlabel('Columnas')
plt.tight_layout()
plt.show()
```

```python
# Gráfica 3: Distribución del monto (rango -500 a 2000 para excluir outliers extremos)
dfventas['Monto_num'] = pd.to_numeric(dfventas['Monto'].str.replace(',', '.'), errors='coerce')
monto_filtrado = dfventas['Monto_num'].clip(-500, 2000)

plt.figure(figsize=(12, 5))
plt.hist(monto_filtrado, bins=60, color='steelblue', edgecolor='white')
plt.axvline(dfventas['Monto_num'].mean(), color='red',   linestyle='--', label=f"Media: ${dfventas['Monto_num'].mean():,.2f}")
plt.axvline(dfventas['Monto_num'].median(), color='orange', linestyle='--', label=f"Mediana: ${dfventas['Monto_num'].median():,.2f}")
plt.axvline(0, color='black', linestyle='-', linewidth=1, label='Monto = 0')
plt.title('Distribución del monto (rango -500 a 2,000)')
plt.xlabel('Monto ($)')
plt.ylabel('Frecuencia')
plt.legend()
plt.tight_layout()
plt.show()
```
