```python
# --2. Limpieza de los datos
# Identificar valores nulos
# Calcular porcentaje de valores faltantes
# Detectar regisitos duplicados
# Aplicar estrategias: eliminacion,transformacion
# Justificar las deciciones tomadas
```

```python
# Identificar valores nulos
df.isnull().sum()
```

```python
# Calcular porcentaje de valores faltantes
(df.isnull().sum() / len(df)) * 100
```

```python
# Detectar registros duplicados
df.duplicated().sum()
```

```python
# Aplicar estrategias: eliminacion,transformacion
df.drop(columns=["director"], inplace=True)
```

```python
# Contar los valores nulos
df.isnull().sum()
```

```python
# Rellenar los valores nulos (NaN) de la columna cast con la moda
df["cast"] = df["cast"].fillna("cast")
```

```python
# Contar los valores nulos
df.isnull().sum()
```

```python
# Rellenar los valores nulos (NaN) de la columna country con la moda
df["country"] = df["country"].fillna("country")
```

```python
# Contar los valores nulos
df.isnull().sum()
```

```python
# Rellenar los valores nulos (NaN) de la columna duration con la moda
df["date_added"] = df["date_added"].fillna(df["date_added"]. mode()[0])
```

```python
# Rellenar los valores nulos (NaN) de la columna rating con la moda
df["rating"] = df["rating"].fillna(df["rating"]. mode()[0])
```

```python
# Contar los valores nulos
df.isnull().sum()
```

```python
# Rellenar los valores nulos (NaN) de la columna duration con la moda
df["duration"] = df["duration"].fillna(df["duration"]. mode()[0])
```

```python
# Contar los valores nulos
df.isnull().sum()
```

```python
# Gráfico de barras (conteo de películas por país)
plt.figure(figsize=(12,6))
sns.countplot(data=df, x="country", order=df["country"].value_counts().head(10).index)
plt.title("Top 10 países con más títulos en Netflix")
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfico de líneas (evolución de lanzamientos por año)
# Agrupar por año y contar títulos
df_year = df.groupby("release_year").size().reset_index(name="count")

plt.figure(figsize=(12,6))
sns.lineplot(data=df_year, x="release_year", y="count")
plt.title("Número de títulos lanzados por año")
plt.xlabel("Año de lanzamiento")
plt.ylabel("Cantidad de títulos")
plt.show()
```

```python
# Histograma de años
plt.figure(figsize=(10,4))
sns.histplot(df['release_year'].dropna(), bins=20, color='#4C72B0')
plt.title('Distribución de títulos por año')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Cantidad')
plt.tight_layout()
plt.show()
```

```python
# Mapa de valores nulos en el dataset
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Mapa de valores nulos en el dataset")
plt.show()
```

```python
# Countplot de clasificaciones (rating)
plt.figure(figsize=(10,5))
top_ratings = df['rating'].value_counts().head(10).index
sns.countplot(data=df, x='rating', order=top_ratings, hue='rating', palette='Set2', legend=False)
plt.title('Top 10 clasificaciones más comunes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
