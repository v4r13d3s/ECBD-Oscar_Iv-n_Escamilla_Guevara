# Parte 4. Selección de variables Predictoras y Variable objetivo

X - Variables que utilizara el modelo
Y - Variable que queremos predecir

```python
columnas = [
    "Pclass",
    "Sex",
    "Age",
    "SibSp",
    "Parch",
    "Fare",
    "Embarked"
]
```

```python
# Definimos x y
x = dfTitanic[columnas]
y = dfTitanic["Survived"]
```

```python
print(x.columns.tolist())
```

```python
print(y.name)
```

## Parte 5. Tratamiento de los valores

Para simplificar el proceso, realizaremos la impuntación de los datos antes de la separación TRIN/TEST

```python
# Revisamos las variables predictoras
x.isnull().sum()
```

```python
x["Age"] = x["Age"].fillna(
    x["Age"].median()
)
```

```python
x["Embarked"] = x["Embarked"].fillna(
    x["Embarked"].mode()[0]
)
```

```python
x.isnull().sum()
```

```python
x.head()
```

## Parte 6. Escalado de los datos

Los algoritmos no entienden texto. Se deben convertir las variables categoricas en variables númericas.

```python
x = pd.get_dummies(
    x,
    columns=["Sex", "Embarked"],
    drop_first=True
)
```

```python
x.head()
```

## Parte 7. Particionamos los datos
No debemos entrenar y evaluar utilizando los mismos datos.

La recomendación es:

Entrenamiento 80% Pruebas 20%

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print("Entrenamiento", x_train.shape)
```

```python
print("Pruebas: ", y_test.shape)
```

## Parte 8. Escalado de las variables

Las variables numericas tienen valores distintos. El escalado evita que una variable domine más que otra

```python
columnas_numericas = [
    "Age",
    "Fare",
    "SibSp",
    "Parch"
]
```

```python
x_train[columnas_numericas].describe()
```

```python
scaler = StandardScaler()
x_train[columnas_numericas] = scaler.fit_transform(
    x_train[columnas_numericas]
)
x_test[columnas_numericas] = scaler.transform(
    x_test[columnas_numericas]
)
```

```python
x_train[columnas_numericas].describe()
```

## Parte 9. Entrenamiento de los datos

Usaremos un arbol de decisión para clasificar registros

```python
modelo = DecisionTreeClassifier (
    max_depth = 4,
    random_state = 42
)
```

```python
modelo.fit(
    x_train,
    y_train
)
```

## Parte 10. Evaluamos el modelo

Realizamos predicciones para evaluar el modelo.

```python
predicciones = modelo.predict(
    x_test
)
```

```python
pd.DataFrame({
    "Real:" : y_test,
    "Prediccion ": predicciones
}).head(30)
```

## Parte 11. Evaluación del modelo

El accuracy determinar el porcentaje de predicciones correctas realizadas por el modelo

```python
accuracy = accuracy_score(
    y_test,
    predicciones
)
```

```python
print ( f"Accuracy :{accuracy: .4f}" )
```

|Accuracy| Interpretacion|

Analizar las variables más relevantes para el algoritmo, al realizar las predicciones
