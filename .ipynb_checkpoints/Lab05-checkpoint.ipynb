```python
#Detectar registros duplicados
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
#Aplicar estrategias de eliminación, transformación.
df.drop(columns = ["rating"], inplace = True)
#Borre la columna rating por que quiero que haya nuevos registros de rating
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
#Rellenamos la column Fare con el valor repetido (moda)
df["fare"] = df["fare"].fillna(df["fare"].mode()[0])
```

```python
#3. Analisis descriptivo (CAPA DE INFORMACIÓN)
#Graficamos el total de pasajeros por clase
plt.figure(figsize=(10,5)) #Tamaño
sns.countplot(df, x="director")
plt.title("Total de directores por pelicula")
plt.xlabel("Clase")
plt.show()
```
